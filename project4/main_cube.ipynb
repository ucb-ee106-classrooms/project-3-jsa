```python
# --- Configuration ---
duration_approach = 3.0 # Approach duration (s)
duration_grasp = 1.5  # Grasp closing duration (s)
duration_lift = 2.5   # Lift duration (s)
framerate = 60      # Simulation/Video framerate

pre_grasp_height_offset = 0.03 # Palm clearance above ball top (m)
lift_height_offset = 0.15      # Vertical lift distance (m)

# --- Determine Indices ---
num_sawyer_actuators = 7
num_allegro_actuators = 16
total_actuators = physics.model.nu
sawyer_actuator_indices = slice(0, num_sawyer_actuators)
hand_actuator_indices = slice(num_sawyer_actuators, num_sawyer_actuators + num_allegro_actuators)

num_ball_qpos = 7 # Free joint qpos (pos + quat)
sawyer_qpos_indices = slice(num_ball_qpos, num_ball_qpos + num_sawyer_actuators)
hand_qpos_indices = slice(num_ball_qpos + num_sawyer_actuators, num_ball_qpos + num_sawyer_actuators + num_allegro_actuators)

# --- Define Constants ---
q_flat_fingers = np.zeros(num_allegro_actuators) # Assumes 0 is flat/open
palm_body_name = 'sawyer/allegro_right/palm'
ball_body_name = 'ball/ball_body'
ball_geom_name = 'ball/ball_geom'

# --- Get Object Info ---
try:
    # Ensure physics state is set to get correct initial positions if needed
    physics.reset()
    set_initial_configuration(physics) # Or a representative pose
    physics.forward()

    ball_body_id = physics.model.body(ball_body_name).id
    ball_geom_id = physics.model.geom(ball_geom_name).id
    ball_initial_pos = physics.data.body(ball_body_name).xpos.copy()
    ball_radius = physics.model.geom(ball_geom_id).size[0]
    palm_body_id = physics.model.body(palm_body_name).id
    print(f"Ball Pos: {ball_initial_pos}, Radius: {ball_radius}")
except KeyError as e:
    raise RuntimeError(f"Model error: Cannot find body/geom needed for setup: {e}")

# --- 1. Grasp Synthesis ---
print("Running Grasp Synthesis...")
physics.reset()
set_initial_configuration(physics) # Use a suitable start state for synthesis
env = AllegroHandEnvCube(physics=physics, q_h_slice=hand_qpos_indices, object_name='ball', cube_center=ball_initial_pos, cube_radius=ball_radius)
q_h0 = physics.data.qpos[hand_qpos_indices].copy()
grasp_fingertip_names = ['sawyer/allegro_right/ff_tip_rubber', 'sawyer/allegro_right/mf_tip_rubber', 'sawyer/allegro_right/rf_tip_rubber', 'sawyer/allegro_right/th_tip_rubber']
q_star = synthesize_grasp(env, q_h0, grasp_fingertip_names) # Returns only hand joint values
if q_star.shape[0] != num_allegro_actuators:
    raise ValueError(f"q_star has wrong shape: {q_star.shape}")
print("Grasp synthesis complete.")

# --- 2. Calculate Key Poses via IK ---
print("Calculating Key Poses...")
ik_params = {'step_size': 0.05, 'tol': 0.01, 'alpha': 0.05, 'damping': 0.05, 'max_steps': 1500,
             'jacp': np.zeros((3, physics.model.nv)), 'jacr': np.zeros((3, physics.model.nv))}

# Helper to run IK
def compute_ik(target_pos, target_orient, target_body_name, start_qpos, physics, ik_params):
    physics.reset()
    physics.data.qpos[:] = start_qpos
    physics.forward()
    # Recreate solver instance with potentially updated physics state reference
    # Pass physics object explicitly to the constructor if required by your IK class
    ik_solver = LevenbergMarquardtIK(physics.model, physics.data, **ik_params, physics=physics)
    q_result_ik = ik_solver.calculate([target_pos], [target_orient], [target_body_name])
    return q_result_ik

# 2a. Initial Pose (arm start + flat fingers)
physics.reset()
set_initial_configuration(physics)
q_initial = physics.data.qpos.copy()
q_initial[hand_qpos_indices] = q_flat_fingers

# 2b. Pre-Grasp Pose (palm @ target height + flat fingers)
pre_grasp_palm_pos = ball_initial_pos + np.array([0, 0, ball_radius + pre_grasp_height_offset])
print(f"Target Pre-Grasp Palm Pos: {pre_grasp_palm_pos}")
q_pre_grasp_ik = compute_ik(pre_grasp_palm_pos, None, palm_body_name, q_initial, physics, ik_params) # Target position only
q_pre_grasp = q_pre_grasp_ik.copy()
q_pre_grasp[hand_qpos_indices] = q_flat_fingers

# 2c. Grasp Pose (arm @ pre-grasp + grasped fingers)
q_grasp = q_pre_grasp.copy()
q_grasp[hand_qpos_indices] = q_star

# 2d. Lift Pose (palm lifted + grasped fingers)
physics.reset(); physics.data.qpos[:] = q_grasp; physics.forward() # Get actual palm pos in grasp state
lift_palm_pos = physics.data.body(palm_body_name).xpos.copy() + np.array([0, 0, lift_height_offset])
print(f"Target Lift Palm Pos: {lift_palm_pos}")
q_lift_ik = compute_ik(lift_palm_pos, None, palm_body_name, q_grasp, physics, ik_params) # Target position only
q_lift = q_lift_ik.copy()
q_lift[hand_qpos_indices] = q_star
print("Finished calculating key poses.")

# --- 3. Simulation Loop ---
steps_approach = int(duration_approach / physics.model.opt.timestep)
steps_grasp = int(duration_grasp / physics.model.opt.timestep)
steps_lift = int(duration_lift / physics.model.opt.timestep)
n_steps = steps_approach + steps_grasp + steps_lift
print(f"Total steps: {n_steps} (Approach: {steps_approach}, Grasp: {steps_grasp}, Lift: {steps_lift})")

# Allocate data
sim_time = np.zeros(n_steps)
ncon = np.zeros(n_steps)
force_on_ball = np.zeros((n_steps, 3)) # World frame force on ball
video = []

# Initialize simulation
physics.reset()
physics.data.qpos[:] = q_initial
physics.data.qvel[:] = 0
physics.forward()

# Simulation loop
render_interval = max(1, int(1 / (framerate * physics.model.opt.timestep)))
options = mj.MjvOption(); mj.mjv_defaultOption(options) # Default viz options
print("Starting simulation...")
with mj.Renderer(physics.model.ptr, 480, 640) as renderer:
    current_step = 0
    # Define phase transitions
    phases = [
        ('approach', steps_approach, q_initial, q_pre_grasp),
        ('grasp',    steps_grasp,    q_pre_grasp, q_grasp),
        ('lift',     steps_lift,     q_grasp, q_lift)
    ]

    for phase_name, num_phase_steps, start_q, end_q in phases:
        print(f"Phase: {phase_name}")
        for i in range(num_phase_steps):
            if current_step >= n_steps: break # Safety break

            alpha = np.clip(i / float(num_phase_steps - 1) if num_phase_steps > 1 else 1.0, 0.0, 1.0)

            # Set Control Targets based on phase
            if phase_name == 'grasp':
                # Interpolate ONLY hand, keep arm fixed at pre-grasp target
                target_hand_q = start_q[hand_qpos_indices] + alpha * (end_q[hand_qpos_indices] - start_q[hand_qpos_indices])
                physics.data.ctrl[sawyer_actuator_indices] = q_pre_grasp[sawyer_qpos_indices]
                physics.data.ctrl[hand_actuator_indices] = target_hand_q
            else: # Approach and Lift
                target_qpos = start_q + alpha * (end_q - start_q)
                physics.data.ctrl[sawyer_actuator_indices] = target_qpos[sawyer_qpos_indices]
                physics.data.ctrl[hand_actuator_indices] = target_qpos[hand_qpos_indices]

            # Step simulation
            physics.step()

            # Record data
            sim_time[current_step] = physics.data.time
            ncon[current_step] = physics.data.ncon
            # Calculate force on ball
            if physics.data.ncon > 0 and ball_geom_id != -1:
                 force_sum = np.zeros(3)
                 for j in range(physics.data.ncon):
                     contact = physics.data.contact[j]
                     if contact.geom1 == ball_geom_id or contact.geom2 == ball_geom_id:
                         contact_frame = contact.frame.reshape(3, 3)
                         force_in_contact_frame = physics.data.contact[j].efc_address # Get forces in EFC coordinates is complex
                         # Alternative: use mj_contactForce per contact (slower but simpler)
                         temp_ft = np.zeros(6)
                         mj.mj_contactForce(physics.model.ptr, physics.data.ptr, j, temp_ft)
                         force_world = contact_frame @ temp_ft[:3]
                         if contact.geom1 == ball_geom_id: force_sum -= force_world
                         else: force_sum += force_world
                 force_on_ball[current_step] = force_sum

            # Render
            if current_step % render_interval == 0:
                renderer.update_scene(physics.data.ptr, "camera_2", options)
                video.append(renderer.render())

            current_step += 1
        if current_step >= n_steps: break # Ensure we don't overshoot buffer
    # End of simulation loop
print(f"Simulation finished. Steps: {current_step}")
```

```python
# --- Plotting ---
if current_step > 0:
    sim_time_actual = sim_time[:current_step]
    force_actual = force_on_ball[:current_step]
    ncon_actual = ncon[:current_step]

    _, ax = plt.subplots(3, 1, sharex=True, figsize=(10, 8)) # Slightly smaller figure
    plt.style.use('seaborn-v0_8-whitegrid') # Cleaner style

    # Plot forces
    ax[0].plot(sim_time_actual, force_actual[:, 0], label='Fx')
    ax[0].plot(sim_time_actual, force_actual[:, 1], label='Fy')
    ax[0].plot(sim_time_actual, force_actual[:, 2], label='Fz')
    ax[0].set_title('Contact Force on Ball (World Frame)')
    ax[0].set_ylabel('Newton')
    ax[0].legend(fontsize='small')

    # Add phase lines
    time_approach_end = steps_approach * physics.model.opt.timestep
    time_grasp_end = time_approach_end + steps_grasp * physics.model.opt.timestep
    for x_val, label, color in [(time_approach_end, 'End Approach', 'r'), (time_grasp_end, 'End Grasp', 'g')]:
       for axis in ax:
           axis.axvline(x_val, color=color, linestyle='--', linewidth=1)
    ax[0].text(time_approach_end, ax[0].get_ylim()[1], ' Approach', color='r', ha='left', va='top')
    ax[0].text(time_grasp_end, ax[0].get_ylim()[1], ' Grasp', color='g', ha='left', va='top')


    # Plot contacts
    ax[1].plot(sim_time_actual, ncon_actual, color='tab:blue')
    ax[1].set_title('Number of Contacts')
    ax[1].set_ylabel('Count')
    max_contacts = int(np.max(ncon_actual)) if ncon_actual.size > 0 else 0
    ax[1].set_yticks(np.arange(0, max_contacts + 2, max(1, max_contacts // 4 + 1))) # Dynamic ticks


    # Plot force magnitude (log)
    force_magnitude = np.linalg.norm(force_actual, axis=1)
    ax[2].plot(sim_time_actual, force_magnitude + 1e-9, color='tab:purple') # Add epsilon for log scale
    ax[2].set_yscale('log')
    ax[2].set_title('Contact Force Magnitude on Ball (Log Scale)')
    ax[2].set_ylabel('Newton')
    ax[2].set_xlabel('Time (s)')

    for axis in ax: axis.grid(True, linestyle=':')
    plt.tight_layout()
    plt.show()
```
